```python
# Two Sum (return indices)
# Problem: Find indices of two numbers that add up to target.

# Example: Input: nums = [2,7,11,15], target = 9 Output: [0,1]

def TwoSum(nums: list[int], target: int ) -> list[int]:
  num_map = {} # Use a dictionary to store numbers and their indices
  for i, num in enumerate(nums):
    complement = target - num
    if complement in num_map: # Check if the complement exists in our map (from previously processed numbers)
      # If found, return the index of the complement and the current number's index
      return [num_map[complement], i]
    # Add the current number and its index to the map for future lookups
    num_map[num] = i
  return [] # Should not be reached if a solution is guaranteed by the problem statement


nums = [2,7,11,15]
target = 9

print(TwoSum(nums, target))
```

```python
# First Non-Repeating Character
# Problem: Return index of first unique character.

# Example: Input: "leetcode" Output: 0


def firstUniqChar( s: str) -> int:
    # Implement your solution here
    rep_freq = {}
    for i,v in enumerate(s):
      rep_freq[v] = rep_freq.get(v, 0 ) + 1

    for i in range(len(s)):
      if rep_freq[s[i]] == 1:
        return i
    return -1




print(firstUniqChar("llettcode"))

```

```python
# Valid Anagram
# Problem: Check if two strings are anagrams.

# Example: Input: "anagram", "nagaram" Output: True

def isAnagram(s: str, t: str) -> bool:
        # Implement your solution here
        if len(s) != len(t) :
          return False
        freq = {}
        for ch in s:
          freq[ch] = freq.get(ch, 0) + 1

        for ch in t:
          if ch not in freq or freq[ch] == 0:
            return False
          freq[ch] -= 1
        return True

print(isAnagram("anagram", "ngaram"))
```
