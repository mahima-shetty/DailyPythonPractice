```python
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]
```

```python

print(a)
a_set = set(a)
```

```python
print(a_set)
```

```python
# Remove all duplicates without using set()
res = []
for x in a:
  if x not in res:
    res.append(x)

print(res)
```

```python
# Find frequency of each element (store in dict)

freq = {}

for x in a :
  freq[x] = freq.get(x,0) + 1

print(freq)
```

```python
# Reverse list without using reverse()

rev = a[::-1]
print(rev)
```

```python
# Reverse list without using reverse()

rev = []
for x in a :
  rev.insert(0, x)

print(rev)
```

```python
# Rotate list right by 2 positions

k = 2
res = a[-k:] + a[:-k]
print(res)

```

```python
# Rotate list left by 3 positions
# 1,2,3,4
# 4123
k = 3
res = a[k:] + a[:k]
print(res)
```

```python
# Find second largest element without sorting
max = 0
sec_max = 0

for x in a:
  if x>max:
    sec_max = max
    max = x
  elif sec_max < x < max:
    sec_max = x

print(sec_max)


```

```python
# Find all pairs whose sum = 10
res = []
for x in a:
  if 10 - x in a:
    res.append([x, 10-x])
    break

print(res)
```

```python
# Flatten:

a = [1, [2,3], [4,[5,6]], 7]


```

```python

def flatten(a):
  res = []
  for x in a:
    if isinstance(x, list):
      res.extend(flatten(x))
    else:
      res.append(x)

  return res


result = flatten(a)
print(result)
```

```python
even = []
odd = []

for x in result:
  if x%2 == 0:
    even.append(x)
  else:
    odd.append(x)

print(even)
print(odd)
```

```python
# Find longest increasing contiguous subarray
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]
max = 1
curr = 1

for i in range(1, len(a)):
  if a[i] > a[i-1]:
    curr += 1
    if curr > max:
      max = curr
  else:
    curr = 1

print(max)
```

```python

# Remove all occurrences of a value (e.g., 5)

k = 5

for i in a:
  if i == k:
    a.remove(i)

print(a)

```

```python

# Find intersection with another list: b = [2,3,10,9]
res = []
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]
b = [1,2,3,4,5]
for i in a:
  if i in b and i not in res:
    res.append(i)

print(res)

```

```python

# Find union without duplicates

for i in a+b:
  if i not in res:
    res.append(i)

print(res)
```

```python

# Count inversions (i < j but a[i] > a[j])

count = 0

for i in range(len(a)):
  for j in range(i+1, len(a)):
    if a[i] > a[j]:
      count += 1

print(count)
```

```python
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]
# Implement bubble sort manually
for i in range(len(a)):
  for j in range(len(a) - i - 1):
    if a[j] > a[j+1]:
      a[j] , a[j+1] = a[j+1] , a[j]
  print(a)


print(a)

```

```python
a = [1, 2, 2, 3, 5, 5, 6, 8, 9, 9]
# Implement binary search (assume sorted list)
left = 0
right = len(a) - 1
mid = (left + right )//2
target = 5
found = False

while left <= right:
  mid = (left + right )//2
  if a[mid] == target:
    print( mid )
    found = True
    break
  elif a[mid] < target:
    left = mid + 1
  elif a[mid] > target:
    right = mid - 1
if not found:
  print('target not found')







```

```python

# Group elements by frequency (high → low)

freq = {}

for x in a:
  freq[x] = freq.get(x, 0) + 1

res = sorted(freq.items(), key = lambda x : x[1], reverse = True )
print(res)

```

```python
# Convert list to dictionary: keys = elements
# values = index positions (list of indices)

res = {}
for i , x in enumerate(a):
  if x not in res:
    res[x] = []
  res[x].append(i)

print(res)

```

```python
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]

# Find subarray with maximum sum (Kadane’s algorithm)


max_sum = a[0]
curr = a[0]


for i in range(1, len(a)):
  curr = __builtins__.max(curr + a[i], a[i] )
  max_sum = __builtins__.max(curr, max_sum)

print(max_sum)
```

```python
# Check if list is palindrome
a = [1,2,3,2,1]
palindrome = True
n = len(a)

for i in range(len(a)//2):
  if a[i] != a[n-i-1]:
    print("Its not palindrome")
    palindrome = False
    break

if palindrome:
  print("Its is palindrome")
```
