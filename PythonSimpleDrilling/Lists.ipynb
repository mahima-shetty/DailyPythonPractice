```python
a = [5, 2, 9, 1, 5, 6, 2, 3, 9, 8]
```

```python

print(a)
a_set = set(a)
```

```python
print(a_set)
```

```python
# Remove all duplicates without using set()
res = []
for x in a:
  if x not in res:
    res.append(x)

print(res)
```

```python
# Find frequency of each element (store in dict)

freq = {}

for x in a :
  freq[x] = freq.get(x,0) + 1

print(freq)
```

```python
# Reverse list without using reverse()

rev = a[::-1]
print(rev)
```

```python
# Reverse list without using reverse()

rev = []
for x in a :
  rev.insert(0, x)

print(rev)
```

```python
# Rotate list right by 2 positions

k = 2
res = a[-k:] + a[:-k]
print(res)

```

```python
# Rotate list left by 3 positions
# 1,2,3,4
# 4123
k = 3
res = a[k:] + a[:k]
print(res)
```

```python
# Find second largest element without sorting
max = 0
sec_max = 0

for x in a:
  if x>max:
    sec_max = max
    max = x
  elif sec_max < x < max:
    sec_max = x

print(sec_max)


```
